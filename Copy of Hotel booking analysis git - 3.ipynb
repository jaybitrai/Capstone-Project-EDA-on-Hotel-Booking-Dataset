# ***How Many Bookings were cancelled***

```python
df['is_canceled'].value_counts()
```

```python
cancelled_bookings = df['is_canceled'].value_counts().iloc[1]
successful_bookings = df['is_canceled'].value_counts().iloc[0]
#0 means not canceled
#1 means canceled
```

```python
#% of bookings canceled
canceled_bookings_ratio = cancelled_bookings/len(df['is_canceled'])*100
canceled_bookings_ratio = round(canceled_bookings_ratio,2)
print(canceled_bookings_ratio)

#% of successful bookings
successful_booking_ratio = 100-canceled_bookings_ratio
print(successful_booking_ratio)
```

```python
# Creating plot
fig = plt.figure(figsize =(10, 7))
mylabel = ['Successful Bookings','Cancelled Bookings']
data = [successful_bookings,cancelled_bookings]
colors = ['green','red']

def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{p:.2f}% ({v:d})'.format(p=pct,v=val)
    return my_autopct

plt.pie(data, labels = mylabel, colors = colors,autopct=make_autopct(data))
plt.legend()

# show plot
plt.show()
```

# ****What is Booking ratio between Resort hotel and city hotel****

```python
#exploring data of hotel types
df['hotel'].value_counts()
```

```python
city_hotel_bookings = df['hotel'].value_counts().iloc[0]
resort_hotel_bookings = df['hotel'].value_counts().iloc[1]
```
