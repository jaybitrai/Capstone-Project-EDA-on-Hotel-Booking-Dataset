# ***How Many Bookings were cancelled***

```python
df['is_canceled'].value_counts()
```

```python
cancelled_bookings = df['is_canceled'].value_counts().iloc[1]
successful_bookings = df['is_canceled'].value_counts().iloc[0]
#0 means not canceled
#1 means canceled
```

```python
#% of bookings canceled
canceled_bookings_ratio = cancelled_bookings/len(df['is_canceled'])*100
canceled_bookings_ratio = round(canceled_bookings_ratio,2)
print(canceled_bookings_ratio)

#% of successful bookings
successful_booking_ratio = 100-canceled_bookings_ratio
print(successful_booking_ratio)
```

```python
# Creating plot
fig = plt.figure(figsize =(10, 7))
mylabel = ['Successful Bookings','Cancelled Bookings']
data = [successful_bookings,cancelled_bookings]
colors = ['green','red']

def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{p:.2f}% ({v:d})'.format(p=pct,v=val)
    return my_autopct

plt.pie(data, labels = mylabel, colors = colors,autopct=make_autopct(data))
plt.legend()

# show plot
plt.show()
```

## **Booking for each year and month**



```python
#exploring data
booking_year_counts = pd.DataFrame(df['arrival_date_year'].value_counts().reset_index())
```

```python
booking_year_counts.rename(columns = {"index":"year","arrival_date_year":"bookings_in_year"},inplace=True)
```

```python
booking_year_counts.sort_values(by=['year'],inplace=True)
```

```python
booking_year_counts
```

```python
#values for plotting graph
years = (list(booking_year_counts['year']))
bookings_every_year = list(booking_year_counts['bookings_in_year'])

fig = plt.figure(figsize = (10,5))

 
# creating the bar plot
plt.plot(years, bookings_every_year, color ='black',linestyle="-.")
plt.bar(years, bookings_every_year, color ='red',width=0.2)



plt.xlabel("year of booking")
plt.ylabel("no of bookings by year")
plt.title("Bookings by year")
plt.show()
```

```python
#drawing_graph
#values for plotting graph
months = (list(booking_month_counts['months']))
bookings_every_month = list(booking_month_counts['arrival_date_month'])

fig = plt.figure(figsize = (10,5))

 
# creating the bar plot
plt.plot(months, bookings_every_month, color ='black',linestyle="-.")
plt.bar(months, bookings_every_month, color ='red',width=0.2)



plt.xlabel("month of booking")
plt.ylabel("no of bookings by month")
plt.title("Months by number of bookings")
plt.show()
```

```python
months_name = ['January','February','March','April','May','June',"July","August","September",'October','November','December']
months_num = [num for num in range(1,13)]
months_dict = dict(zip(months_name,months_num))
```

```python
months_number = pd.DataFrame([months_dict[a] for a in list(booking_month_counts['months'])])
```

```python
booking_month_counts = pd.concat([booking_month_counts,months_number],axis=1)
```

```python
booking_month_counts.rename(columns={0:'months_number',"arrival_date_month":"total_bookings_in_month"},inplace=True)
```

```python
booking_month_counts.sort_values(by=['months_number'],inplace=True)
```

```python
booking_month_counts.drop(['months_number'],axis=1,inplace=True)
```

```python
booking_month_counts.reset_index()
```

```python
booking_month_counts
```

```python
#drawing_graph
#values for plotting graph
months = (list(booking_month_counts['months']))
bookings_every_month = list(booking_month_counts['total_bookings_in_month'])

fig = plt.figure(figsize = (10,5))

 
# creating the bar plot
plt.plot(months, bookings_every_month, color ='black',linestyle="-.")
plt.bar(months, bookings_every_month, color ='red',width=0.2)



plt.xlabel("month of booking")
plt.ylabel("no of bookings by month")
plt.title("Bookings by month")
plt.show()
```
